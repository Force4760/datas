```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
marcador_medicoes = [0.49, 0.53, 0.52, 0.51,0.49]
marcador = np.mean(marcador_medicoes)
ma = 259.89e-3
mb1 = 254.98e-3
mb2 = 754.98e-3
```

```python
def get_v(x,t):
    return x/t
def get_p(m, v):
    return m*v
def get_ec(m, v):
    return m*(v*v)/2
def linear(m,x, b):
    return m*x +b
```

```python
data = {"ta (ms)": [5.767, 3.256, 8.280, 10.683, 12.715],
        "tab (ms)": [12.186, 18.996, 18.373, 24.029, 32.011]}
df = pd.DataFrame(data, columns = ["ta (ms)", "tab (ms)"])
df["va (m/s)"] = get_v(marcador * 10**-2,df["ta (ms)"]*10**-3) 
df["vab (m/s)"] = get_v(marcador * 10**-2,df["tab (ms)"]*10**-3)

df["pi (kgms⁻¹)"] = get_p(ma,df["va (m/s)"]) 
df["pf (kgms⁻¹)"] = get_p(ma+mb1,df["vab (m/s)"]) 

df["Eci (kgms⁻¹)"] = get_ec(ma,df["va (m/s)"]) 
df["Ecf (kgms⁻¹)"] = get_ec(ma+mb1,df["vab (m/s)"]) 
df
```

```python
data2 = {"ta (ms)": [3.993, 8.204, 12.988, 10.301, 14.247],
        "tab (ms)": [19.296, 34.575, 63.571, 43.260, 63.729]}
df2 = pd.DataFrame(data2, columns = ["ta (ms)", "tab (ms)"])
df2["va (m/s)"] = get_v(marcador * 10**-2,df2["ta (ms)"]*10**-3) 
df2["vab (m/s)"] = get_v(marcador * 10**-2,df2["tab (ms)"]*10**-3)

df2["pi (kgms⁻¹)"] = get_p(ma,df2["va (m/s)"]) 
df2["pf (kgms⁻¹)"] = get_p(ma+mb2,df2["vab (m/s)"]) 

df2["Eci (kgms⁻¹)"] = get_ec(ma,df2["va (m/s)"]) 
df2["Ecf (kgms⁻¹)"] = get_ec(ma+mb2,df2["vab (m/s)"]) 
df2
```

```python
data3 = {"ti (ms)": [11.884, 9.745, 9.192, 8.277, 7.336],
        "tf (ms)": [14.467, 11.096, 10.206, 9.207, 8.376]}
df3 = pd.DataFrame(data3, columns = ["ti (ms)", "tf (ms)"])
df3["vi (m/s)"] = get_v(marcador * 10**-2,df3["ti (ms)"]*10**-3) 
df3["vf (m/s)"] = get_v(marcador * 10**-2,df3["tf (ms)"]*10**-3)
df3
```

```python
data4 = {"ti (ms)": [10.237, 9.730, 6.372, 8.694, 12.276],
        "tf (ms)": [21.392, 19.897, 16.645, 17.923, 24.620]}
df4 = pd.DataFrame(data4, columns = ["ti (ms)", "tf (ms)"])
df4["vi (m/s)"] = get_v(marcador * 10**-2,df4["ti (ms)"]*10**-3) 
df4["vf (m/s)"] = get_v(marcador * 10**-2,df4["tf (ms)"]*10**-3)
df4
```

```python
x1 = df3["vi (m/s)"]
y1 = df3["vf (m/s)"]

plt.grid()
t1 = plt.scatter(x1,y1)

plt.xlabel(r"$V_{inicial} (m/s)$")
plt.ylabel(r"$V_{final} (m/s)$")

m,b = np.polyfit(x1, y1, 1) 
m = int(m * 10000)/10000
b = int(b * 10000)/10000
plt.plot(x1,linear(m,x1,b), label=(r'$V_f = $'+ str(m) +r"$x $" + str(b)))
plt.legend(loc='upper left')
```
